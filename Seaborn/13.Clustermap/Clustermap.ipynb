 
<p>Seaborn’s Clustermap function is great for making simple heatmaps and hierarchically-clustered heatmaps with dendrograms on both rows and/or columns.

Most often when you try to make heatmap you would also like to cluster row-wise or col-wise to see if there is any pattern emerges. Seaborn’s Clustermap’s ability to hierarchically-cluster exactly solves that problem.
Its just Plot a matrix dataset as a hierarchically-clustered heatmap.
</p>

```python
# Plot a clustered heatmap:

import seaborn as sns

sns.set(color_codes=True)

iris = sns.load_dataset("iris")
iris.head(5)
```

```python
species = iris.pop("species")
iris.head()
```

```python
 sns.clustermap(iris)
```

```python
# Change the size and layout of the figure:

sns.clustermap(iris, figsize=(15, 12),row_cluster=False,
               
                   dendrogram_ratio=(.1, .2),cbar_pos=(0, .2, .03, .4))
```

```python
# Add colored labels to identify observations:

lut = dict(zip(species.unique(), "rbg"))
row_colors = species.map(lut)
sns.clustermap(iris, row_colors=row_colors)
```

```python
# Use a different colormap and adjust the limits of the color range:

sns.clustermap(iris, cmap="mako", vmin=0, vmax=10)
```

```python
# Use a different similarity metric:

sns.clustermap(iris, metric="correlation")
```

```python
# Use a different clustering method:

sns.clustermap(iris, method="single")
```

```python
# Standardize the data within the columns:

sns.clustermap(iris, standard_scale=1)
```

```python
# Normalize the data within the rows:

sns.clustermap(iris, z_score=0, cmap="vlag")
```
